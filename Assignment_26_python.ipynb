```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
""" scattar plot is plot in which data is plotted into form of some dots """
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.title("this is title")
plt.plot(x,y)
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
"""The subplot() function in Matplotlib is used to create a grid of subplots within a single figure"""

import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.plot([0, 1, 2, 3, 4, 5], [0, 100, 200, 300, 400, 500])
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot([0, 1, 2, 3, 4, 5], [50, 20, 40, 20, 60, 70])
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot([0, 1, 2, 3, 4, 5],[10, 20, 30, 40, 50, 60])
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot([0, 1, 2, 3, 4, 5], [200, 350, 250, 550, 450, 150])
plt.title('Plot 4')

plt.tight_layout()

plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
""" 
A bar plot, also known as a bar chart, 
is a graphical representation of categorical data using rectangular bars."""
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
"""A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. 
It displays key statistical measures such as the median, quartiles, and potential outliers."""
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
```
